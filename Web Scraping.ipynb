```python
#Making dataframe For all the Columns(Final)

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Find the pagination elements
pagination_elements = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, '//ul[@class="pagination"]/li/a')))

# Exclude the "Next" button from the total number of pages
total_pages = int(pagination_elements[-2].text)

# Initialize empty lists to store scraped data
case_subjects = []
case_numbers = []
case_titles = []
author_judges = []
upload_dates = []
judge_dates = []
citations = []
sccitations = []
downloads = []
pdf_urls = []

# Iterate over each page
for page in range(1, total_pages + 1):
    # Click on the page number
    page_number_element = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, f'//ul[@class="pagination"]/li/a[text()="{page}"]')))
    driver.execute_script("arguments[0].click();", page_number_element)
    
    # Wait for the table to load
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, '//tbody')))
    
    # Scrape data from the current page
    case_subjects += [subject.text for subject in driver.find_elements(By.XPATH, '//tbody//td[2]')]
    case_numbers += [number.text for number in driver.find_elements(By.XPATH, '//tbody//td[3]')]
    case_titles += [title.text for title in driver.find_elements(By.XPATH, '//tbody//td[4]')]
    author_judges += [judge.text for judge in driver.find_elements(By.XPATH, '//tbody//td[5]')]
    upload_dates += [upload.text for upload in driver.find_elements(By.XPATH, '//tbody//td[6]')]
    judge_dates += [jdate.text for jdate in driver.find_elements(By.XPATH, '//tbody//td[7]')]
    citations += [cit.text for cit in driver.find_elements(By.XPATH, '//tbody//td[8]')]
    sccitations += [sc_cit.text for sc_cit in driver.find_elements(By.XPATH, '//tbody//td[9]')]
#     downloads += [dl.find_element(By.TAG_NAME, 'a').text if dl.find_elements(By.TAG_NAME, 'a') else '' for dl in driver.find_elements(By.XPATH, '//tbody//td[10]')[:len(case_subjects)]]
    
    pdf_links = driver.find_elements(By.XPATH, '//tbody//td[10]//a')
    pdf_urls += [link.get_attribute('href') for link in pdf_links[:len(case_subjects)]]
    
#     # Print the length of lists after scraping each page
#     print(f"Page {page} - Subject: {len(case_subjects)}, Number: {len(case_numbers)}, Title: {len(case_titles)}, Author Judge: {len(author_judges)}, Upload Date: {len(upload_dates)}, Judge Date: {len(judge_dates)}, Citation: {len(citations)}, SC Citation: {len(sccitations)}, Download: {len(downloads)}, PDF URL: {len(pdf_urls)}")

# Adjust the length of pdf_urls to match the other lists
if len(pdf_urls) < len(case_subjects):
    pdf_urls += [None] * (len(case_subjects) - len(pdf_urls))

# Create a DataFrame from the scraped data
data = {
    'Case_Subject': case_subjects,
    'Case_Number': case_numbers,
    'Case_Title': case_titles,
    'Author Judge': author_judges,
    'Upload Date': upload_dates,
    'Judge Date': judge_dates,
    'Citation': citations,
    'SC Citation': sccitations,
#     'Download': pdf_urls
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
# Find the pagination elements
pagination_elements = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, '//ul[@class="pagination"]/li/a')))

# Exclude the "Next" button from the total number of pages
total_pages = int(pagination_elements[-2].text)

# Initialize empty lists to store scraped data
downloads = []
pdf_urls = []

# Iterate over each page
for page in range(1, total_pages + 1):
    # Click on the page number
    page_number_element = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, f'//ul[@class="pagination"]/li/a[text()="{page}"]')))
    driver.execute_script("arguments[0].click();", page_number_element)
    
    # Wait for the table to load
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, '//tbody')))
    
    # Scrape data from the current page
    downloads += [dl.find_element(By.TAG_NAME, 'a').text if dl.find_elements(By.TAG_NAME, 'a') else '' for dl in driver.find_elements(By.XPATH, '//tbody//td[10]')[:len(case_subjects)]]
    pdf_links = driver.find_elements(By.XPATH, '//tbody//td[10]//a')
    pdf_urls += [link.get_attribute('href') for link in pdf_links[:len(case_subjects)]]

# Adjust the length of pdf_urls to match the other lists
if len(pdf_urls) < len(downloads):
    pdf_urls += [None] * (len(downloads) - len(pdf_urls))

# Create a DataFrame from the scraped data
data = {
    'Download': pdf_urls
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
